## Cleaning Functions

```python
def clean_screen_size(val):
    if not val or val == 'NeedToBeFilled':
        return None
    val = val.lower().replace('"', '').replace("'", '').replace('inch', '').replace('pouce', '').strip()
    val = val.replace(',', '.')
    try:
        match = re.search(r'\d+(\.\d+)?', val)
        if match:
            num = float(match.group())
            if 100 <= num < 200: num = num / 10.0
            if 9 <= num <= 20:
                return round(num, 1)
            return None
    except:
        pass
    return None

def clean_frequency(val):
    if not val or val == 'NeedToBeFilled':
        return None
    val = val.lower().replace('hz', '').strip()
    try:
        return int(float(val))
    except:
        return None

def clean_resolution(val):
    if not val or val == 'NeedToBeFilled':
        return "Unknown"
    val_upper = val.upper()
    if 'FHD' in val_upper or '1920X1080' in val_upper: return '1920x1080'
    if '1366X768' in val_upper or ('HD' in val_upper and 'FHD' not in val_upper): return '1366x768'
    if 'QHD' in val_upper or '2560X1440' in val_upper: return '2560x1440'
    if 'UHD' in val_upper or '4K' in val_upper: return '3840x2160'
    match = re.search(r'(\d{3,4})[xX](\d{3,4})', val)
    if match: return f"{match.group(1)}x{match.group(2)}"
    return val

def clean_city(val):
    if not val or val == 'NeedToBeFilled':
        return "Unknown"
    val = val.upper().strip()
    if val == 'EZZOUAR': return 'BAB EZZOUAR'
    return val
```
